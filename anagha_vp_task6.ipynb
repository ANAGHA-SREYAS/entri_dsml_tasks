##4.  Write a Python program to plot two or more lines with legends, different widths and colors. The code snippet gives the output shown in the following screenshot:



image.png

````python
import matplotlib.pyplot as plt
import numpy as np
x1=np.array([10,20,30])
y1=np.array([20,40,10])
x2=np.array([10,20,30])
y2=np.array([40,10,30])
plt.plot(x1,y1,color='blue',linewidth=3,label='line-1-width-3')
plt.plot(x2,y2,color='red',linewidth=5,label='line-2-width-5')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Two or more lines with different width and colors with suitable legends')
plt.legend(loc='upper right')
plt.xlim(min(min(x1), min(x2)), max(max(x1), max(x2)))
plt.ylim(min(min(y1), min(y2)), max(max(y1), max(y2)))
plt.show()
````

##Exercise 1: Read Total profit of all months and show it using a line plot


````python
import matplotlib.pyplot as plt
import pandas as pd
df=pd.read_csv('company_sales_data.csv')
x=df['month_number']
y=df['total_profit']
plt.plot(x,y)
plt.xlabel('Month number')
plt.ylabel('Profit in dollar')
plt.xticks(x)
plt.title('Company profit per month')
plt.yticks([100000, 200000, 300000, 400000, 500000])
plt.show()


````

##Exercise 8: Calculate total sale data for last year for each product and show it using a Pie chart
Note: In Pie chart display Number of units sold per year for each product in percentage.

The Pie chart should look like this.



image.png

````python
import matplotlib.pyplot as plt
import pandas as pd
df=pd.read_csv('company_sales_data.csv')
my_labels=['facecream','facewash','toothpaste','bathingsoap','shampoo','moisturizer']
salesData   = [df ['facecream'].sum(), df ['facewash'].sum(), df ['toothpaste'].sum(),
         df ['bathingsoap'].sum(), df ['shampoo'].sum(), df ['moisturizer'].sum()]
plt.figure(figsize=(10,8))
plt.pie(salesData,labels=my_labels,autopct='%1.1f%%')
plt.legend(loc='lower right')
plt.title('Sales data')
plt.show()
````

##Exercise Question 10: Read all product sales data and show it using the stack plot
The Stack plot should look like this.

image.png


````python
import matplotlib.pyplot as plt
import pandas as pd


import matplotlib.pyplot as plt
import pandas as pd
df=pd.read_csv('company_sales_data.csv')
months = df['month_number']
facecream_sales = df['facecream']
facewash_sales = df['facewash']
toothpaste_sales = df['toothpaste']
bathingsoap_sales = df['bathingsoap']
shampoo_sales = df['shampoo']
moisturizer_sales = df['moisturizer']
plt.figure(figsize=(10,6))

plt.stackplot(months, facecream_sales, facewash_sales, toothpaste_sales, bathingsoap_sales, shampoo_sales, moisturizer_sales,
              labels=['Face Cream', 'Face Wash', 'Toothpaste', 'Bathing Soap', 'Shampoo', 'Moisturizer'],
              colors=['violet','grey','red','black','green','yellow'])
plt.xlabel('Month Number')
plt.ylabel('Sales units in Number')
plt.title('All Product Sales Data')
plt.legend(loc='upper left')
plt.show()


````
